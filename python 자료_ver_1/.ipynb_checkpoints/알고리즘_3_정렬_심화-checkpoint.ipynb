정렬이란 데이터를 특정한 조건에 따라 일정한 순서가 되도록 다시 배열하는 일

### 1. 버블정렬
버블 정렬(Bubble sort)은 이웃하는 숫자를 비교하여 작은 수를 앞쪽으로 이동시키는 과정을 반복하여 정렬한다

1. 첫 번째 패스


2. 두 번째 패스


3. 세 번째 패스


```python
list_data = [95, 80, 70, 90, 40, 35, 75, 65, 45, 20, 15, 10]

for p in range(len(list_data)-1, 0, -1):
    for i in range(p):
        if list_data[i] > list_data[i+1]:
            list_data[i], list_data[i+1] = list_data[i+1], list_data[i]
                
print(list_data)
```

[문제] 입력이 [3, 4, 7, 2, 1]일 때 버블 정렬 알고리즘의 3번째 패스가 종료된 결과는?

- [1,2,3,4,7]

### 2. 칵테일 정렬
버블 정렬에서 가장 큰 수는 토끼같이 빠르게 아래쪽으로 내려가서 제자리를 잡지만, 작은 수들은 기껏해야 한 칸씩 거북이처럼 위로 올라간다.
칵테일 정렬은 작은 수들에게도 토끼가 될 기회를 주자는 것이다.

칵테일 정렬은 홀수 번째 패스에서 버블 정렬처럼 위에서 아래 방향으로 이웃한 숫자들을 비교하며 위에 있는 숫자가 크면 서로 교환한다. 그리고 짝수 번째 패스에서는 아래에서 위 방향으로 이웃한 숫자를 비교하여 아래에 있는 숫자가 작으면 서로 교환한다. 
이는 마치 칵테일을 만들 때 위 아래로 흔드는 것과 같다고 해서 정렬 이름을 칵테일 정렬이라 부른다.




[문제] 입력이 [8, 6, 2, 5, 3, 9, 1, 7]일 때 칵테일 정렬 알고리즘의 3번째 패스가 종료된 결과는?

```python
list_data = [95, 80, 70, 90, 40, 35, 75, 65, 45, 20, 15, 10]
length = len(list_data)
top = 0
bottom = length - 1

swapped = True

while swapped:
    swapped = False
    for i in range(top, bottom):
        if list_data[i] > list_data[i+1]:
            list_data[i], list_data[i+1] = list_data[i+1], list_data[i]
            swapped = True
    
    if not swapped:
        break
    
    bottom -= 1
    swapped = False
    for i in range(bottom -1, top - 1, -1):
        if list_data[i] > list_data[i + 1]:
            list_data[i], list_data[i + 1] = list_data[i + 1], list_data[i]
            swapped = True
            
    top += 1
    
print(list_data)
```

### 3. 삽입 정렬
삽입 정렬(Insertion sort)은 나열된 숫자들의 앞부분은 정렬된 상태를 유지하면서 정렬이 안 된 뒷부분의 가장 왼쪽 숫자를 앞부분의 적절한 위치에 삽입하여 정렬되도록 하는 과정을 반복한다.

[1] 두번째 숫자를 현재 원소로 하여 시작한다.
[2] 현재 원소를 왼쪽 방향의 숫자와 하나 씩 비교하여 현재 원소가 삽입될 곳을 찾아서 그 곳에 삽입한다.
[3] 이제 새 현재 원소는 직전 현재 원소가 있었던 곳의 바로 오른쪽 옆의 숫자이며, [2]를 수행한다. 

[문제] 입력이 a = [9, 5, 8, 4, 7, 3, 6, 2]일 때 삽입 정렬 알고리즘이 3번째 현재 원소(즉, a[3])를 앞부분에 삽입한 결과는?

```python
list_data = [95, 80, 70, 90, 40, 35, 75, 65, 45, 20, 15, 10]

for i in range(1, len(list_data)):
    for j in range(i, 0, -1):
        if list_data[j-1] > list_data[j]:
            list_data[j], list_data[j-1] = list_data[j-1], list_data[j]
            
print(list_data)
```
